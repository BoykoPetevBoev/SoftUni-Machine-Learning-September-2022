```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.datasets import make_blobs, make_moons, make_circles
from sklearn.metrics import silhouette_samples
from scipy.cluster.hierarchy import linkage, dendrogram
```

# Clustering Demo

## KMeans with Blob

```python
attributes, clusters = make_blobs()
plt.scatter(attributes[:, 0], attributes[:, 1], c=clusters)
plt.show()
```

```python
k_mean = KMeans(n_clusters=3, init="random")
k_mean.fit(attributes)
predictions = k_mean.predict(attributes)
```

```python
plt.scatter(attributes[:, 0], attributes[:, 1], c=predictions)
plt.show()
```

## KMeans with Blob Mixed 

```python
attributes_mixed, clusters_mixed = make_blobs(centers=[[1, 1], [2, 2.5], [1.1, 1.2]], cluster_std=0.5)
```

```python
plt.scatter(attributes_mixed[:, 0], attributes_mixed[:, 1], c=clusters_mixed)
plt.show()
```

```python
k_mean_mixed = KMeans(n_clusters=3, init="random")
k_mean_mixed.fit(attributes_mixed)
predictions_mixed = k_mean_mixed.predict(attributes_mixed)
```

```python
plt.scatter(attributes_mixed[:, 0], attributes_mixed[:, 1], c=predictions_mixed)
plt.show()
```

## KMeans with Moons

```python
moons_attributes, moons_cluster = make_moons(noise=0.07)
```

```python
plt.scatter(moons_attributes[:, 0], moons_attributes[:, 1], c=moons_cluster)
plt.show()
```

```python
k_means_moons = KMeans(n_clusters=2)
k_means_moons.fit(moons_attributes)
moons_predictions = k_means_moons.predict(moons_attributes)
```

```python
plt.scatter(moons_attributes[:, 0], moons_attributes[:, 1], c=moons_predictions)
plt.show()
```

## KMeans with Circles

```python
circles_attributes, circles_cluster = make_circles(factor=0.5, noise=0.02)
```

```python
plt.scatter(circles_attributes[:, 0], circles_attributes[:, 1], c=circles_cluster)
plt.show()
```

```python
k_means_circles = KMeans(n_clusters=2)
k_means_circles.fit(circles_attributes)
circles_predictions = k_means_circles.predict(circles_attributes)
```

```python
plt.scatter(circles_attributes[:, 0], circles_attributes[:, 1], c=circles_predictions)
plt.show()
```
